```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
mainpath = "datasets"
filename = "data_ingresante.xls"
fullpath = os.path.join(mainpath, filename)
```

```python
data = pd.read_excel(fullpath, sheet_name='data')
```

```python
data.head()
```

## ¿Cuál es el porcentaje de fracaso de académico de los estudiantes ingresantes 2019, 2020 en su primer año de estudio y segundo año de estudio? 

```python
num_2019_1 = data[(data['PeriodoIngresoAnio'] == 2019) & (data['Desertor'] == 1) & (data['Anio'] == 2019) & (data['Edad'] > 16) ]
total_2019_1 = data[(data['PeriodoIngresoAnio'] == 2019 ) & ( data['Anio'] == 2019 ) & (data['Edad'] > 16)]

porcentaje_1 = len(num_2019_1)*100/len(total_2019_1)
print("Porcentaje de Fracaso Académico en Ingresantes 2019 => ", porcentaje_1, "%")
```

```python
num_2019_1.groupby("ProgramaAcademico")[["Desertor","Edad", "Peso","Talla","IMC","NotaPromedioPonderado","PuntajeFichaEvalSocioEconomico"]].mean()
```

## ¿Cuántos estudiantes ingresantes 2019, 2020 desaprobaron más del 50%, 70 % de sus asignaturas desaprobadas? 

```python
estudiantes_50_2019 = data[(data['PeriodoIngresoAnio'] == 2019) & ((data['TotalAsignaturaDesAprobado'] * 100 / data['TotalAsignatura'] ) > 50)]
estudiantes_70_2019 = data[(data['PeriodoIngresoAnio'] == 2019) & ((data['TotalAsignaturaDesAprobado'] * 100 / data['TotalAsignatura'] ) > 70)]

estudiantes_50_2020 = data[(data['PeriodoIngresoAnio'] == 2020) & ((data['TotalAsignaturaDesAprobado'] * 100 / data['TotalAsignatura'] ) > 50)]
estudiantes_70_2020 = data[(data['PeriodoIngresoAnio'] == 2020) & ((data['TotalAsignaturaDesAprobado'] * 100 / data['TotalAsignatura'] ) > 70)]
```

```python
print("Estudiantes Desaprobados Ingresantes 2019 más del 50% => ", len(estudiantes_50_2019), " estudiantes")
print("Estudiantes Desaprobados Ingresantes 2019 más del 70% => ", len(estudiantes_70_2019), " estudiantes")

print("Estudiantes Desaprobados Ingresantes 2020 más del 50% => ", len(estudiantes_50_2020), " estudiantes")
print("Estudiantes Desaprobados Ingresantes 2020 más del 70% => ", len(estudiantes_70_2020), " estudiantes")
```

```python
estudiantes_50_2019.groupby("ProgramaAcademico")[["Desertor","Edad", "Peso","Talla","IMC","NotaPromedioPonderado","PuntajeFichaEvalSocioEconomico"]].mean()
```
